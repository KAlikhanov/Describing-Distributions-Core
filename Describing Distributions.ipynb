```python
# Importing Libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Importing data.
df = pd.read_csv('data(1).csv')
df.head()
```

## Visualizations:

```python
# VitD_levels
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10,10))
sns.histplot(x=df['VitD_levels'],kde=True,ax=ax)
ax.set_title('Vitamin D Levels Histogram')

mean = np.mean(df['VitD_levels'])
median = np.median(df['VitD_levels'])
std = df['VitD_levels'].std()
plus_one_std = mean + std
minus_one_std = mean - std

ax.axvline(mean, ls='--',lw=3, label=f"Vitamin D Levels (Mean={mean:.2f})",color='Red')
ax.axvline(median, ls='--', lw=3, label=f"Vitamin D Levels (Median={median:.2f})",color = 'Green')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0, alpha = 0.4)

plt.legend()
plt.tight_layout()
```

 Vitamin D Levels:
 - This data is continuous.
 - This is a normal distribution so no skew.
 - I believe this is mesokurtic.

```python
# Doc_visits
fig, ax = plt.subplots(figsize=(10,10))
sns.histplot(x=df['Doc_visits'],kde=True,ax=ax)
ax.set_title('Doctor Visits Histogram')

mean = np.mean(df['Doc_visits'])
median = np.median(df['Doc_visits'])
std = df['Doc_visits'].std()
plus_one_std = mean + std
minus_one_std = mean - std

ax.axvline(mean, ls='--',lw=3, label=f"Doctor Visits (Mean={mean:.2f})",color='Red')
ax.axvline(median, ls='--', lw=3, label=f"Doctor Visits (Median={median:.2f})",color = 'Green')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0, alpha = 0.4)

plt.legend()
plt.tight_layout()
```

Doctor Visits:
- This is discrete data.
- Looks normal so no skew.
- I would probably say this  is leptokurtic.
